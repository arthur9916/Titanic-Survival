# Exploratory Data Analysis

```python
column_target=['Survived']
columns_train=['Age','Pclass','Sex','Fare']
X=train[columns_train]
Y=train[column_target]

X_2=test[columns_train]
```

```python
X['Age']=X['Age'].fillna(X['Age'].median())
```

```python
d={'male':0,'female':1}
X['Sex']=X['Sex'].apply(lambda x:d[x])
X['Sex'].head()
```

```python
d={'male':0,'female':1}
X_2['Sex']=X_2['Sex'].apply(lambda x:d[x])
X_2['Sex'].head()
```

```python
def age_init(age):
    if age<18.0:
        return "young"
    elif age>=40.0:
        return "old"
    else:
        return "median"
    
```

```python
X["Age"]=X["Age"].apply(age_init)
X["Age"].head(10)
```

```python
X_2["Age"]=X_2["Age"].apply(age_init)
X_2["Age"].head(10)
```

```python
d={'young':0,'median':1,'old':2}
X['Age']=X['Age'].apply(lambda x:d[x])
X['Age'].head(10)
```

```python
from sklearn.metrics import f1_score,accuracy_score
```

```python
prediction=accuracy_score(Y_pred,Y_test)
prediction
```
